```python
#Group the data on continent and the population From 1970-2022
df2= df.groupby("Continent")[df.columns[5:13][::-1]].mean().sort_values(by="2022 Population", ascending=False)
df2
```

```python
df3 = df2.transpose()
df3
```

```python
#Plot the transposed dataframe of grouped data
df3.plot()
```
